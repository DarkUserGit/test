Hello, be welcome to the **<span style="color:#009EE3">OpenSignals Tools</span>** environment !!!  

This page is intended to make available to the users of *Plux Wireless Biosignals* systems, such as **bitalino** or **biosignalsplux**, a set of **<span class="color5">Jupyter Notebooks</span>**.    

In these informative and interactive resources it is explained and demonstrated how some processing operations can be done in **<span class="color1">Python</span>**.
For example, it is explained how the acquired signals can be loaded and ploted.
Additionally, are presented some detection methods of ECG R peaks and muscular activation periods on EMG data.  

Many of the processing operations described in **<span class="color5">Jupyter Notebooks</span>** are incorporated in *opensignalstools* library.  

This **<span class="color1">Python</span>** package contains functions specially defined to answer the needs of the **<span style="color:#009EE3">OpenSignals</span>** users.

For an integrated, automatic and broad processing experience, using multiple processing operations simultaneously, the user can purchase **<span style="color:#009EE3">OpenSignals</span>** plugins.
